```python
# Dependencies 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Files to Load 
alldata_path = "Resources/alldata.csv"
DataScientist_path = "Resources/DataScientist.csv"
```

```python
# The correct encoding must be used to read the CSV in pandas
dfa = pd.read_csv(alldata_path)
dfb = pd.read_csv(DataScientist_path)
```

```python
#importing the dataset
dfa = pd.read_csv('alldata.csv')
dfa
```

```python
#importing the dataset
dfb = pd.read_csv('DataScientist.csv')
dfb
```

```python
# Delete extraneous column
del dfa['Description']
dfa.head()
```

```python
# Identify incomplete rows in alldata
dfa.count()
```

```python
# Drop all rows with missing information in alldata
dfa = dfa.dropna(how='any')
```

```python
# Verify dropped rows in alldata
dfa.count()
```

```python
# Delete extraneous column in DataScientist csv
dfb.drop(['Job Description', 'index', 'Headquarters', 'Size', 'Founded', 'Type of ownership', 'Revenue', 'Competitors', 'Easy Apply'], axis=1, inplace=True)
dfb.head()
```

```python
# Identify incomplete rows in DataScientist csv
dfb.count()
```

```python
# Drop all rows with missing information in alldata
dfb = dfb.dropna(how='any')
```

```python
# Verify dropped rows in alldata
dfb.count()
```

```python
# Merge both dataframes using an outer join
merge_df = pd.merge(dfa, dfb, on="Location", how="outer")
merge_df
```
